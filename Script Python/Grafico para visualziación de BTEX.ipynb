```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

### Grafico para visualziación de BTEX
# Data
variables = ['Benceno', 'Tolueno', 'Ethylbenzeno', 'Xyleno']
values = [30, 40, 45, 35]

# Number of variables
num_vars = len(variables)

# Compute angle for each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# The plot is a circle, so we need to "complete the loop" and append the start to the end
values += values[:1]
angles += angles[:1]

# Create figure
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Draw one axe per variable + add labels
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

plt.xticks(angles[:-1], variables)

# Draw ylabels
ax.set_rscale('log')

# Plot data
ax.plot(angles, values, linewidth=1, linestyle='solid')

# Fill area
ax.fill(angles, values, 'b', alpha=0.1)

plt.show()


### Grafico para visualización de temperatura

# Data
temperature = 22
min_temp = -10
max_temp = 50

# Create figure and axes
fig, ax = plt.subplots(figsize=(2, 6))  # Adjust figsize for vertical orientation

# Create the temperature bar
ax.bar(0, temperature, width=0.5, bottom=min_temp, color='red')

# Set axis limits and labels
ax.set_ylim(min_temp, max_temp)
ax.set_xlim(-1, 1)  # Adjust x-axis limits for bar width
ax.set_xticks([])  # Hide x-axis ticks
ax.set_ylabel('Temperatura (°C)')
ax.set_title('Temperatura')

# Add background colors to represent different temperature ranges
ax.axhspan(min_temp, 0, color='lightblue', alpha=0.3)
ax.axhspan(0, 20, color='lightgreen', alpha=0.3)
ax.axhspan(20, 30, color='yellow', alpha=0.3)
ax.axhspan(30, max_temp, color='red', alpha=0.3)

# Add temperature label
ax.text(0, temperature + 1, f'{temperature}°C', ha='center', va='bottom')

plt.show()
#plot a air velocty meter
import plotly.graph_objects as go

fig = go.Figure(go.Indicator(
    mode = "gauge+number",
    value = 270,
    title = {'text': "Velocidad de viento"},
    gauge = {'axis': {'range': [0, 500]},
             'steps': [
                 {'range': [0, 250], 'color': "lightgray"},
                 {'range': [250, 400], 'color': "gray"}],
             'threshold': {'line': {'color': "red", 'width': 4}, 'thickness': 0.75, 'value': 490}}))

fig.show()









```
